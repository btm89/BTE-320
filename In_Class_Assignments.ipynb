```python
Sales = float(input("Enter Sales Value ($)"))
Tax_Rate = float(input("Enter Tax_Rate (%)"))
COGS = (0.4 * Sales)
Gross_Profit = Sales - COGS
Taxes = Gross_Profit * Tax_Rate/100
Net_Profit = Gross_Profit - Taxes
Net_Income = Gross_Profit - Taxes

print(f"Sales: {Sales}$")
print(f"COGS: {COGS}$")
print(f"Gross_Profit: {Gross_Profit}$")
print(f"Taxes: {Taxes}$")
print(f"Net_Income: {Net_Income}$")
```
